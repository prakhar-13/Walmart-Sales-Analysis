#### Task 1:- Which Month has the Highest Sales

```python
df.Month
```

```python
Monthly_Sales=df.groupby(['Month']).sum(numeric_only=True)
Monthly_Sales
```

```python
months=range(1,13)
sales=Monthly_Sales['Weekly_Sales']
```

```python
plt.bar(months,sales);
plt.xticks(months);
plt.xlabel('Month');
plt.ylabel('Sales');
```

#### From the above plot we can see that 'July' has the highest sales

### Task 3:-Which store had the highest sales

```python
df.Store.unique()
```

###### There are 45 store from 1 to 45

```python
store_sales=df.groupby(['Store']).sum(numeric_only=True)
store_sales
```

```python
stores=range(1,46)
```

```python
plt.plot(stores,store_sales['Weekly_Sales']);
plt.xticks(stores, rotation='vertical', size=7);
plt.grid()
```

#### From the above plot we can observe that store"20" had the highest sales and store"4" had the second highest sales
